```r
tmp <- ls(paste("package", "BSgenome.Ecoli.NCBI.20080805", sep=":"))
```

```r
genome <- eval(parse(text=tmp))
```

```r
genome
```

```r
ecoli <- getSeq(genome)
```

```r
names(ecoli) <- seqnames(genome)
```

```r
writeXStringSet(ecoli, file="ecoli.fasta", format="fasta", width=50)
```

```r
library(Biostrings)
```

```r
ecoli <- readDNAStringSet("ecoli.fasta", format="fasta")
```

```r
snf_ecoli <- alphabetFrequency(ecoli)
```

```r
snf_ecoli
```

```r
tmp <- cbind(names(ecoli), snf_ecoli)
```

```r
write.table(tmp, "snf_ecoli.txt", sep="\t", append=F, quote=F, row.names=F, col.names=T)
```
