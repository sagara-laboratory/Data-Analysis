```r
x <- seq(-7, 7, 0.1)
```

```r
curve(dnorm(x, 0, 0.5), from=-7, to=7,  type="l")
```

```r
curve(dnorm(x, 0, 2), from=-7, to=7, type="l", col="blue")
```

```r
curve(dnorm(x, -2, 0.8), from=-7, to=7, type="l",col="red")
```

```r
curve(dnorm(x, 0, 0.5), from=-7, to=7,  type="l")
curve(dnorm(x, 0, 2), col="blue", add=TRUE)
curve(dnorm(x, -2, 0.8), col="red", add=TRUE)
```

```r
x <- seq(-10, 10, 0.1)
plot(x, dnorm(x, 0, 1), xlim=c(-10,10),  type="h", ylab="")
```

```r
x <- seq(-10, 10, 0.5)
plot(x, dnorm(x, 2, 1), col="blue", xlim=c(-10,10),  type="h", ylab="")
```

```r
x <- seq(-10, 10, 1)
plot(x, dnorm(x, -2, 1), col="red", xlim=c(-10,10),  type="h", ylab="")
```

```r
x <- seq(-10, 10, 0.1)
plot(x, dnorm(x, 0, 1), xlim=c(-10,10),  type="h", ylab="")
par(new="T")
x <- seq(-10, 10, 0.5)
plot(x, dnorm(x, 2, 1), col="blue", xlim=c(-10,10),  type="h", ylab="")
par(new="T")
x <- seq(-10, 10, 1)
plot(x, dnorm(x, -2, 1), col="red", xlim=c(-10,10),  type="h", ylab="")
```
