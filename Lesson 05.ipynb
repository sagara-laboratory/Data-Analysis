```r
plot.new()
x = 0:10
p <- seq(0.1, 0.9, 0.1)
y <- sapply(p, function(a){dbinom(x, 10, a)})
matplot(x, y, type="o", lty=1, ylim=c(0, 1), pch=1:9, col=1:9, ylab="density", main="Binomial distribution")
legend(2, 1.1, paste("p=", p, "(n=10)", sep=""), col=1:9, lty=1, bty="n")
abline(h=0, v=0, lty=2, col=2)
```

```r
plot.new()
x = 0:25
n <- c(5, 10, 20, 30, 50)
y<-sapply(n, function(a){dbinom(x, a, 0.2)})
matplot(x, y, type="o", lty=1, ylim=c(0,0.45), pch=1:5, col=1:5, ylab="density", main="Binomial distribution")
legend(5, 0.4, paste("n=", n, "(p=0.2)", sep=""), col=1:5, lty=1, bty="n")
abline(h=0, v=0, lty=2, col=2)
```

```r
plot.new()
 x = 0:20
 n <- c(10, 20, 30, 50)
 p <- c(0.6, 0.3, 0.2, 0.12)
 y <- matrix(0, nrow=21, ncol=4)
 for(j in 1:4){
	d <- dbinom(x, n[j], p[j])
	y[,j] <- d
 }
 matplot(x, y, type="o", lty=1, ylim=c(0, 0.3), pch=1:4, col=1:4, ylab="density",  main="Binomial distribution")
 legend(10, 0.3, paste("n=", n, "p=", p,sep=""), col=1:4, lty=1, bty="n")
 abline(h=0, v=0, lty=2, col=2)
```
